#Handle categorical data

first lets see the features with datatype integers

```python
df_int= df.select_dtypes(int)
df_int
```

we have 16 columns with dtypes integer.

now lets find out features with dtypes object. we need to convert the object to integer.

```python
df_cat= df.select_dtypes(object)
df_cat
```

we  observe that we have 17 columns with dtypes object.

we use labelEncoder to change categorical value to integer. 

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in df_cat:
   df_cat[col]=le.fit_transform(df_cat[col])

```

```python
df_cat.head()
```

now all categorical datas are converted to numerical.

lets join two table df_cat (this contains columns with categorical values) and df_int (this contains columns with integer value.). we combine both of them to form single table with all the numerical values.  

```python
df=pd.concat([df_cat, df_int], axis=1, join='inner')
df
```

now all the features in dataset has numerical value. 

#Split x and y

seperate our features and target.

#chi square test for feature selection

we can perform chi square test to select import features for our model. for this first lets split our data to train and test. based on the output of this test we can know the priority of our feature. 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=40, test_size=0.2)
```

```python
from sklearn.feature_selection import chi2
ch_score= chi2(x_train, y_train)
ch_score
```

The first row are values of pvalue and 2nd of zvalue

arrange the feature in ascending order of pvalue

```python
pvalues=pd.Series(ch_score[1])
pvalues.index= x_train.columns
pvalues.sort_values(ascending=True)
```

arrange the feature with descending order of zvalue (more the zvalue more significant is feature)

```python
z=pd.Series(ch_score[0])
z.index= x_train.columns
z.sort_values(ascending=False)
```

we have sorted our features on the basis of pvalues. smaller the pvalue more significant the feature. so we choose feature based on these values.

#feature select

```python
subdf=df[['G2', 'G1', 'absences', 'failures', 'Dalc', 'school', 'Walc', 'reason', 'Mjob', 'Medu','Fedu', 'studytime', 'schoolsup', 'health', 'sex']]
```

on the basis of pvalue we have choosed first fifteen significant feature.

```python
x=subdf.copy()
y=df.G3
```

#split data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(x, y, random_state = 40, test_size = 0.2)
```

#fit and evalute model

#linear regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_squared_error(y_pred, y_test)
```

```python
mean_absolute_error(y_pred, y_test)
```

#Ridge Regression

```python
from sklearn.linear_model import Ridge
model= Ridge(alpha=1.0)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
mean_squared_error(y_pred, y_test)
```

```python
mean_absolute_error(y_pred, y_test)
```

#lasso regression

```python
from sklearn import linear_model
model= linear_model.Lasso(alpha=0.1)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
mean_squared_error(y_pred, y_test)
```

```python
mean_absolute_error(y_pred, y_test)
```

#Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
model = Pipeline([('poly', PolynomialFeatures(degree=2)),
                ('linear', LinearRegression())])
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
mean_squared_error(y_pred, y_test)
```

```python
mean_absolute_error(y_pred, y_test)
```
